```python
#Importing important packages and modules for the web scrapping
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import time
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
#Creating an empty list
linka=[]
```

```python
#Calling WebDriver to extract all the links from "href" tag and append it to the empty list
driver=webdriver.Chrome("chromedriver.exe")
cities=['ahmedabad','bangalore','chennai','delhi-ncr','gurgaon','hyderabad','jaipur','kolkata','mumbai','new-delhi','noida','pune']
for i in cities:
    url="https://www.cardekho.com/used-cars+in+"+i
    driver.get(url)
    time.sleep(30)
# previous_height=driver.execute_script("return document.body.scrollHeight")
    ''' while True:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
        time.sleep(3)
        new_height=driver.execute_script("return document.body.scrollHeight")
        if new_height==previous_height:
            break '''
    elem = driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]//a[@href]')
    for i in elem:
        x=i.get_attribute("href")
        linka.append(x)
```
