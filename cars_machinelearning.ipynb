```python
#Let us extract the data of "company name" & "name of car" from the  "Name" column of the dataset using a python code.
company=[]
name_car=[]
for i in df.index:
    x=df.loc[i,'Name'].split(" ")
    y=x[1]
    z=x[2:]
    company.append(y)
    name_car.append(z)
    
```

```python
#Lets clean the "Number of owners" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute.
for i in df.index:
    df.loc[i,"Number of owners"]=int(df.loc[i,"Number of owners"][0])
```

```python
#Lets clean the "KMs Driven" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    df.loc[i,"KMs Driven"]=int(df.loc[i,"KMs Driven"].split(" ")[0].split(",")[0]+df.loc[i,"KMs Driven"].split(" ")[0].split(",")[1])
```

```python
#Lets clean the "Engine Displacement" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    df.loc[i,"Engine Displacement"]=int(df.loc[i,"Engine Displacement"].split(" ")[0])
```

```python
#Lets clean the "Engine Capacity" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    df.loc[i,"Engine Capacity"]=int(df.loc[i,"Engine Capacity"].split(" ")[0])
```

```python
#Lets clean the "Mileage" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    df.loc[i,"Mileage"]=float(df.loc[i,"Mileage"].split(" ")[0])
```

```python
#Lets clean the "Maximum Power" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    if "b" in df.loc[i,"Maximum Power"]:
        df.loc[i,"Maximum Power"]=(df.loc[i,"Maximum Power"].split("b")[0])
    elif "at" in df.loc[i,"Maximum Power"]:
        df.loc[i,"Maximum Power"]= (df.loc[i,"Maximum Power"].split(" ")[0])
    else:
        df.loc[i,"Maximum Power"]=df.loc[i,"Maximum Power"]
        
```

```python
#There is only one row present in the dataset where the Torque is in kgm unit (11.4kgm=111.8Nm),let us convert it into Nm unit.
df.loc[159,"Torque"]="111.8Nm"
```

```python
#Lets clean the "Torque" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    if "n" in df.loc[i,"Torque"]:
            df.loc[i,"Torque"]=float(df.loc[i,"Torque"].split("n")[0])
    elif "N" in df.loc[i,"Torque"]:
        df.loc[i,"Torque"]=float(df.loc[i,"Torque"].split("N")[0])
```

```python
#Lets clean the "Price" column in order to make it fully numerical ,so that our final machine learning model can understand this attribute
for i in df.index:
    df.loc[i,"Price"]=float(df.loc[i,"Price"].split(",")[0]+df.loc[i,"Price"].split(",")[1]+df.loc[i,"Price"].split(",")[2])
```

```python
#Let us drop the "Engine Displacement column " since this column is similar to engine capacity column present in the dataset
df=df.drop(['Engine Displacement'], axis = 1)
```

```python
#We are dropping the "Name" column since we have extracted the data from it and transformed the data into two new column which are "Name of Car Company" & "Name of Car".
df=df.drop(['Name'], axis = 1)
```

```python
#Replacing the Null values present in "Maximum Power" & "Torque" with the median values of these repective columns.
df['Maximum Power'].fillna(df['Maximum Power'].median(), inplace=True)
df['Torque'].fillna(df['Torque'].median(), inplace=True)
```

```python
#Bifurcating the training dataset into dependent and independent variables.
x=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
x.shape
```

```python
y
```

```python
#splitting the dataset for train ,test and split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
#Performing standard scaling on the dataset.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train = pd.DataFrame(sc.fit_transform(x_train),columns = x_train.columns)
x_test = pd.DataFrame(sc.transform(x_test),columns = x_test.columns)


#y_train = pd.DataFrame(sc.fit_transform(y_train))


#y_test = pd.DataFrame(sc.transform(y_test))
```

```python
#converting the independent variable y to numpy array .
y_train=y_train.to_numpy()
y_train=y_train.reshape(-1,1)

y_test=y_test.to_numpy()
y_test=y_test.reshape(-1,1)
```

```python
#Importing different regression metrics and machine learning algorithms.
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, train_test_split,GridSearchCV
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
#Assigning two variables (n-number of samples ,k-number of independent features) for the calculation of adjusted r2 score.
n=812
k=34
```

```python
#Linear Regression model
model="Linear Regression Model"
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
Adj_r2 = 1-(1-r2_score(y_test,y_pred))*(n-1)/(n-k-1)
print(f"*************R2 SCORE OF {model}*******************")
print(f"R2_score of {model} is :", r2_score(y_test,y_pred))
print(f"************* ADJUSTED R2 SCORE OF {model}*******************")
print(f"Adjusted_r2_score  of {model} is : ",Adj_r2)
print(f"*************MEAN SQUARED ERROR {model}*******************")
print(f"Mean_squared_error of {model} is : ",mean_squared_error(y_test,y_pred))
```

```python
#DecisionTreeRegressor model
model="Decision Tree Regressor Model"
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred=dtr.predict(x_test)
Adj_r2 = 1-(1-r2_score(y_test,y_pred))*(n-1)/(n-k-1)
print(f"*************R2 SCORE OF {model}*******************")
print(f"R2_score of {model} is :", r2_score(y_test,y_pred))
print(f"************* ADJUSTED R2 SCORE OF {model}*******************")
print(f"Adjusted_r2_score  of {model} is : ",Adj_r2)
print(f"*************MEAN SQUARED ERROR {model}*******************")
print(f"Mean_squared_error of {model} is : ",mean_squared_error(y_test,y_pred))
```

```python
#knearest Neighbours regressor model
knr=KNeighborsRegressor()
model="KNeighbours Regressor Model"
knr.fit(x_train,y_train)
y_pred=knr.predict(x_test)
Adj_r2 = 1-(1-r2_score(y_test,y_pred))*(n-1)/(n-k-1)
print(f"*************R2 SCORE OF {model}*******************")
print(f"R2_score of {model} is :", r2_score(y_test,y_pred))
print(f"************* ADJUSTED R2 SCORE OF {model}*******************")
print(f"Adjusted_r2_score  of {model} is : ",Adj_r2)
print(f"*************MEAN SQUARED ERROR {model}*******************")
print(f"Mean_squared_error of {model} is : ",mean_squared_error(y_test,y_pred))
```

```python
#AdaBoost Regressor model
abr=AdaBoostRegressor()
model="AdaBoost Regressor Model"
abr.fit(x_train,y_train)
y_pred=abr.predict(x_test)
Adj_r2 = 1-(1-r2_score(y_test,y_pred))*(n-1)/(n-k-1)
print(f"*************R2 SCORE OF {model}*******************")
print(f"R2_score of {model} is :", r2_score(y_test,y_pred))
print(f"************* ADJUSTED R2 SCORE OF {model}*******************")
print(f"Adjusted_r2_score  of {model} is : ",Adj_r2)
print(f"*************MEAN SQUARED ERROR {model}*******************")
print(f"Mean_squared_error of {model} is : ",mean_squared_error(y_test,y_pred))
```
